```python
import pandas as pd
```

```python
data =  pd.read_csv(r'C:\Users\clozy\opencv\Scripts\jpnb\CREATING\classification\1.0\gender_voice_dataset.csv')

data.sample(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
data['label'] = labelEncoder.fit_transform(data['label'].astype(str))

data.head()
```

```python
data.boxplot(by = 'label', column=['meanfreq'], grid = False, figsize=(10, 8))
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('label', axis=1)
Y = data['label']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(penalty='l2', solver='liblinear')
logistic_model.fit(x_train, y_train)
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
#visualizing classification results
from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, y_pred)

print(confusion_matrix)
```

```python
print('Training score : ', logistic_model.score(x_train, y_train))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

#no of voices correctly classified
print('Accuracy : ', acc)
#of all  voices identified as male(+ identification in data) how many classified correctly 
print('Precision score : ', pre)
#of all male voices in dataset how many classified correctly 
print('Recall score : ', recall)
```
